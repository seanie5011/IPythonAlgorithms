## Gauss Seidel Method

The Gauss Seidel Method is used to solve a set of linear equations with iterative techniques, it is an extension of the Jacobi Method.

Remember, if we look at each individual equation (each row), we can see that it is possible to rearrange to have one of the unknowns on the left, with the others added with their inputs and the corresponding output on the right:

\begin{equation} x_i = \frac{1}{a_{ii}} \left[ b_i - \sum\limits_{j = 1, j \neq i}^{n} a_{ij}x_j \right] \end{equation}

Computed for each unknown $i$. Essentially, this means that we can use the previous guesses of the unknowns to update each new guess, simply by rearranging the equation and iterating the above.

However, notice that even when we compute a new guess for a given unknown, it is not used in computing the following in guesses for each unknown after that in the sequence. Instead, let's use the newly updated guess for a given unknown when calculating the new guesses for the unknowns after that one:

\begin{equation} x_i = \frac{1}{a_{ii}} \left[ b_i - \sum\limits_{j = 1}^{i - 1} a_{ij}x_{j}^{(k)} - \sum\limits_{j = i + 1}^{n} a_{ij}x_{j}^{(k - 1)} \right] \end{equation}

*Note: The power of $(k)$ and $(k - 1)$ denote which guess we are using, $k$ is the newly updated, while $k - 1$ is the old guess*

*Also note: The limits of the summations reflect which guesses we are using, everything after $i$ we have not yet computed a new guess for*

* Create an initial set of guess
  * If none given, default to all zeroes
  
* Update each guess by the above equation, factoring in the newly updated guesses

* Check if the new guesses are close to the old guesses
  * Solution found if so
  
* End once maximum iterations reached, or if new guesses are sufficiently close to old guesses

Let us now create this in Python.

First we must define a matrix set of linear equations, for simplicity I will use what was given in L11.

\begin{bmatrix}
10 & -1 & 2 & 0 & 6 \\
-1 & 11 & -1 & 3 & 25 \\
2 & -1 & 10 & -1 & -11 \\
0 & 3 & -1 & 8 & 15
\end{bmatrix}

```python
import numpy as np

a = np.array([[10, -1, 2, 0], 
              [-1, 11, -1, 3], 
              [2, -1, 10, -1], 
              [0, 3, -1, 8]])

b = np.array([[6],
              [25],
              [-11],
              [15]])
```

Now let us define the algorithm, we need to pass in the Matrix of inputs $A$, the outputs $b$, the Tolerance (how accurate we want to be) $TOL$, the maximum number of iterations $N$, and optionally the initial guess $x0$.

Note that there are comments intended to help understanding, and be sure to refer to the pseudocode if stuck!

```python
def GaussSeidelMethod(a, b, TOL, N, x0 = None):
    A = np.copy(a)
    # Get size n
    n = len(A)
    # Prepare vectors x and x0
    x = np.zeros([n, 1])
    if x0 is None:
        x0 = np.zeros([n, 1])
    else:
        # Make sure x0 contains floats
        x0 = x0.astype('float64')
    
    # Step 1
    k = 1
    
    # Step 2
    while k <= N:
        # Step 3
        for i in range(n):
            summation_old = 0
            summation_new = 0
            for j in range(n):
                if j < i:
                    summation_new += A[i][j] * x[j] # using newly computed guesses
                if j > i:
                    summation_old += A[i][j] * x0[j] # using old guesses
            
            x[i] = (-summation_new - summation_old + b[i]) / A[i][i]
            
        # Step 4
        if np.linalg.norm(x - x0) < TOL: # gets the L2 norm of x - x0
            print("Tolerance reached.")
            return x
        
        # Step 5
        k += 1
        
        # Step 6
        x0 = x.copy()
        
    # Step 7
    print("Maximum Iterations reached.")
    return x
```

We can now test this algorithm for a solution, which will be returned as an array x:

```python
GaussSeidelMethod(a, b, 0.0000000001, 100)
```
